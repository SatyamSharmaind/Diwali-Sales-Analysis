### EDA

### Gender

```python
data.columns
```

```python
gender_counts = data['Gender'].value_counts()

```

```python
plt.bar(gender_counts.index, gender_counts.values, color=['skyblue','pink'])
plt.xlabel('Gender')
plt.ylabel('Count')
plt.title('Count of Gender (M vs F)')
plt.show()
```

```python
ax=sns.countplot(x='Gender', data=data, palette=['skyblue','pink'])
for bars in ax.containers:
    ax.bar_label(bars)
plt.title('Count of Gender (M vs F)')
plt.show()
```

```python
ax = sns.countplot(x='Gender', data=data)
for bars in ax.containers:
    ax.bar_label(bars)
```

```python
Sales_gen=data.groupby(["Gender"],as_index=False)["Amount"].sum().sort_values(by="Amount",ascending=False)
```

```python
sns.barplot(x="Gender",y="Amount",data=Sales_gen,palette=["orange","skyblue"])
```

### Age

```python
data.columns
```

```python
ax=sns.countplot(x="Age Group",data=data,hue="Gender")
```

```python
ax=sns.countplot(x="Age Group",data=data,hue="Gender")
for bars in ax.containers:
    ax.bar_label(bars)
```

```python
Sales_age=data.groupby(["Age Group"],as_index=False)["Amount"].sum().sort_values(by="Amount",ascending=False)
```

```python
sns.barplot(x="Age Group",y="Amount",data=Sales_age,hue="Age Group")
```

### State

```python
data.columns
```

```python
Sales_state=data.groupby(["State"],as_index=False)["Orders"].sum().sort_values(by="Orders",ascending=False).head(10)
sns.barplot(x="State",y="Orders",data=Sales_state,hue="State")
sns.set(rc={"figure.figsize":(15,3)})
plt.show()
```

```python
Sales_state=data.groupby(["State"],as_index=False)["Amount"].sum().sort_values(by="Amount",ascending=False).head(10)
sns.barplot(x="State",y="Amount",data=Sales_state,hue="State")
sns.set(rc={"figure.figsize":(16,3)})
plt.show()
```

## Martial status

```python
ax=sns.countplot(data=data,x="Marital_Status")
sns.set(rc={"figure.figsize":(7,5)})
for bars in ax.containers:
    ax.bar_label(bars)
```

```python
Sales_Marital=data.groupby(["Marital_Status","Gender"],as_index=False)["Amount"].sum().sort_values(by="Amount",ascending=False)
sns.barplot(x="Marital_Status",y="Amount",data=Sales_Marital,hue="Gender")
sns.set(rc={"figure.figsize":(6,5)})
```

### Occupation

```python
ax=sns.countplot(data=data,x="Occupation")
sns.set(rc={"figure.figsize":(20,5)})
for bars in ax.containers:
    ax.bar_label(bars)
```
